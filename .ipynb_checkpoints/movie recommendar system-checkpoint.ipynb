```python
import re

def clean_tags(text):
    # Remove extra spaces between characters (e.g., "S c i - F i" -> "Sci-Fi")
    text = ''.join(text.split())
    # Then, add space before each capitalized word (for broken joined words like JamesCameron)
    text = re.sub(r'([a-z])([A-Z])', r'\1 \2', text)
    return text.lower()

new_df['tags'] = new_df['tags'].apply(clean_tags)
```

```python
import re

def clean_tags(text):
    # Replace commas, dots, and slashes with spaces
    text = re.sub(r'[,./]', ' ', text)

    # Add space between lowercase and uppercase (e.g., JamesCameron -> James Cameron)
    text = re.sub(r'([a-z])([A-Z])', r'\1 \2', text)

    # Add space between digits and letters (e.g., 22ndCentury -> 22 nd Century)
    text = re.sub(r'(\d)([a-zA-Z])', r'\1 \2', text)
    text = re.sub(r'([a-zA-Z])(\d)', r'\1 \2', text)

    # Lowercase and remove multiple spaces
    text = text.lower()
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
new_df['tags'] = new_df['tags'].apply(clean_tags)
```
